```python
# Create prime numbers between 0 and 100
def prime_nos():
    prime = []
    for i in range(0,101):
        if i>1:
            for j in range(2, i):
                if i%j==0:
                    break
            else:
                prime.append(i)
    return prime
```

```python
# sum nos
sum(prime_nos())
```

```python
''.join([str(i) for i in prime_nos()])
```

```python
def dict_words(x):
    words = x.lower().split()
    first = set([i[0] for i in words])
    dict_words = {}
    for i in first:
        dict_words[i] = set([w for w in words if w[0]==i])
    return dict_words
```

```python
#create a pig translator
#words that start with voewel =+ 'ay'
#move consonant to the end
#treat y as vowel

def pig_latin(x):
    x = x.lower()
    pig = []
    for word in x.split():
        if word[0] in ['a', 'e', 'i', 'o', 'u', 'y']:
            pig.append("{}ay".format(word))
        else:
            while word[0] not in ['a', 'e', 'i', 'o', 'u', 'y']:
                word = word[1:] + word[0]
            pig.append("{}ay".format(word))
    return ' '.join(pig)
```

```python
x = 'i love the elegance of python'
```

```python
pig_latin(x)
```

```python
def foo(word, formatting):
    formatting = list(formatting)
    for i in range(0,len(word),2):
        if word[i]!='run':
            formatting[i]='x'
    for i in range(1,len(word),2):
        if word[i]!='jump':
            formatting[i]='/'
    return ''.join(formatting)
```

```python
foo(['run', 'run', 'jump', 'jump', 'run'], '_|_|_')
```
