```python
df = pd.read_csv("delivery_orders_march.csv")
```

```python
df.head()
```

```python
df['origin'] = df.selleraddress.apply(lambda x: x.split()[-1].lower())
```

```python
df['dest'] = df.buyeraddress.apply(lambda x: x.split()[-1].lower())
```

```python
df['SLA'] = 7
df.loc[np.where((df.dest=='luzon') & (df.origin.isin(['luzon', 'manila'])), True, False), 'SLA'] = 5
df.loc[np.where((df.origin=='manila') & (df.dest.isin(['manila'])),True, False), 'SLA'] = 3
```

```python
df.SLA.value_counts()
```

```python
#convert to gmt+8

df['pick_2'] = df['pick'].apply(lambda x: datetime.datetime.fromtimestamp(x, tz= pytz.timezone('Asia/Singapore')).date())
df['1st_deliver_attempt_2'] = df['1st_deliver_attempt'].apply(lambda x: datetime.datetime.fromtimestamp(x, tz= pytz.timezone('Asia/Singapore')).date())

df.loc[df['2nd_deliver_attempt'].isna(), '2nd_deliver_attempt_2'] = np.datetime64('NaT')
df.loc[~df['2nd_deliver_attempt'].isna(), '2nd_deliver_attempt_2'] = df.loc[~df['2nd_deliver_attempt'].isna(), '2nd_deliver_attempt'].apply(lambda x: datetime.datetime.fromtimestamp(x, tz= pytz.timezone('Asia/Singapore')).date())
```

```python
HOLIDAYS = pd.Series([datetime.date(2020, 3, 8), datetime.date(2020, 3, 25), datetime.date(2020, 3, 30), datetime.date(2020, 3, 31)]).values.astype('datetime64[D]')

def calc_bdays(samp, col1, col2):
    return np.busday_count(samp[col1].values.astype('datetime64[D]'),samp[col2].values.astype('datetime64[D]'),weekmask='1111110',holidays=HOLIDAYS)
```

```python
df['1st_bdays'] = calc_bdays(df, 'pick_2', '1st_deliver_attempt_2')
df.loc[df['2nd_deliver_attempt'].notna(),'2nd_bdays'] = calc_bdays(df.loc[df['2nd_deliver_attempt'].notna()], '1st_deliver_attempt_2', '2nd_deliver_attempt_2')
```

## Holidays and working days

```python
pd.options.display.max_rows = 500
```

```python
df['first_late'] = np.where(df['1st_bdays'] > df['SLA'], 1, 0)
df['second_late'] = np.where(df['2nd_bdays'] > 3, 1, 0)
df['is_late'] = np.where((df.first_late==1) | (df.second_late==1), 1, 0)
```
